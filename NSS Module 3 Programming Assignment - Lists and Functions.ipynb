```python
#9.2 Define a generator function called get_odds() that returns the odd numbers from range(10).
#Use a for loop to find and print the third value returned.

def get_odds():
    for x in range(10):
        if x % 2 ==1:
            yield x

count = 0
for value in get_odds():
    count += 1
    if count == 3:
        print(value)
        break
```
