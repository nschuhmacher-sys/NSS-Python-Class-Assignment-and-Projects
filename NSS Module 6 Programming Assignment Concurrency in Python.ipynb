```python
#15.1 Use multiprocessing to create three separate processes. Make each one wait a random number 
#of seconds between zero and one, print the current time, and then exit.
import multiprocessing
import time
import random
from datetime import datetime
import os

def process_task(process_id):
    wait_time = random.random()
    time.sleep(wait_time)
    current_time = datetime.now()
    print(f"Process {process_id} (PID {os.getpid()}) finished waiting at: {current_time} after {wait_time:.4f} seconds.")

if __name__ == "__main__":
    processes = []
    for i in range(3):
        p = multiprocessing.Process(target=process_task, args=(i,))
        processes.append(p)
        p.start()
    for p in processes:
        p.join()
#this version works but does not print anything in jupyter
```
